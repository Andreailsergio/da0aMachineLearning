```python
import matplotlib.pyplot as plt
plt.plot([1,4,-2,7])
plt.ylabel("y")
plt.xlabel("ascissa")
plt.title("TITOLO GRAFICO")
plt.show()
# ascisse predefinite da programma (0.0, 1.0, 2.0, 3.0)
```

```python
# qui inserisco io i valori delle ascisse
plt.plot([0,2,7,40],[1,4,-2,7])
plt.ylabel("y")
plt.xlabel("ascissa")
plt.title("TITOLO GRAFICO")
plt.show()
```

```python
plt.plot([0,2,4,7],[-2,6,-5,8],'go--')
plt.plot([0,2,5,7],[2,3,5,8],'r-')
# colore linea b=blu r=rosso g=verde -- tratteggiata o con pallini o- punti con linea continua
plt.show()
```

```python
# scatterplot: rappresenta una quantità rispetto ad un'altra
d = {
    'a':[1,2,4],
    'b':[3,5,1],
    's':[50,100,1500]
}
plt.scatter('a','b',s='s',data=d) 
# a ascisse, b ordinate, s è la dimensione (size) dei punti, data sono i parametri che gli passo (un dictionary)
plt.show
```

```python
categorie=['A','B'] #nomi barre istogramma
valori=[10,50]
size=[20,5000]
plt.figure(figsize=(10,3)) # x ampliare la dimensione del riquadro

plt.subplot(121) 
# come in matlab -> quante righe: 1 quante colonne: 2 i grafici che sto per disegnare dove vanno? nella 1° cella
plt.bar(categorie,valori)

plt.subplot(122)
plt.scatter(categorie,valori,s=size)

plt.suptitle("TITOLO GRAFICO")

plt.show()
```

## prendo dati su valore DOGEcoin da https://www.kaggle.com/sudalairajkumar/cryptocurrencypricehistory

```python
import csv
with open("03coin_Dogecoin.csv") as f:
    reader = csv.reader(f,delimiter=',')
    for riga in reader:
        print(riga[7])
        # a noi interessa solo il campo 'close' - valore alla chiusura
```

```python
# lettura dati
dati =  []
with open("03coin_Dogecoin.csv") as f:
    reader = csv.reader(f,delimiter=',')
    for riga in reader:
        dati.append(riga[7])
# cleanup dati
dati=dati[1:] # prendo dalla seconda riga in poi (tolgo intestazione 'close')
dati = [float(x) for x in dati] # trucco: ciclo dentro la lista trasformando una a una le righe da testo in float

#rappresentazione
plt.figure(figsize=(20,5))
plt.title("andamento quotazioni dogecoin")
plt.plot(dati)
plt.ylabel("quotazione (USD)")
plt.show()
```
