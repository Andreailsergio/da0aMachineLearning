### lettura e analisi di un file CSV reale da Internet

[dataset](https://corgis-edu.github.io/corgis/csv/earthquakes/)


```python
import csv

dataE=[]
# INTESTAZIONI ['id,"impact.gap","impact.magnitude","impact.significance","location.depth","location.distance","location.full","location.latitude","location.longitude","location.name","time.day","time.epoch","time.full","time.hour","time.minute","time.month","time.second","time.year"']
# DATI ['nc72666881,"122.0","1.43","31","15.12","0.1034","13km E of Livermore, California","37.6723333","-121.619","California","27","1469593183550","2016-07-27 00:19:43","0","19","7","43","2016"']
def parse_row(row):
    #pass # = per ora, cara funzione, non fare nulla
    # dataE.append(row)
    dataE.append({
        'id':row[0],
        'impact':int(row[3]),
        'location':row[6],
        'state':row[9],
        'year':row[17]
    })
# nella prima riga (la 0) di earthquakes.csv ci sono le intestazioni
with open("earthquakes.csv") as f:
    reader = csv.reader(f,delimiter=',')
    for pos,riga in enumerate(reader):
        #print(riga)
        if pos>0:
            parse_row(riga)
print(f"sono presenti {len(dataE)} record\n")
#print(dataE)
```

```python
# estrarre elenco degli impact
impacts = []
for d in dataE:
    impacts.append(d['impact'])
print(impacts,"\n")
print("IMPATTO MEDIO: ",sum(impacts)/len(impacts))
```

```python
# ANALISI DATI es. calcolo mediana
def calcolo_mediana(lunghezza):
    if lunghezza%2==0:
        mediana_index = int(lunghezza/2)
        print("mediana (con n° records pari): ", impacts[mediana_index])
    else:
        mediana_index1 = int(lunghezza/2)
        mediana_index2 = mediana_index1+1
        print("mediana (con n° records dispari): ", int((impacts[mediana_index1]+impacts[mediana_index2])/2))

impacts.sort()
calcolo_mediana(len(impacts))

impacts.pop()
calcolo_mediana(len(impacts))
```

```python
import requests
import json
r = requests.get("https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_hour.geojson")
print(r.text,"\n")
# print(r.json())
dataset = r.json()
features = dataset['features']
#print(features)
print(json.dumps(features, sort_keys='id', indent=5))
```

```python
data = []
# JSON ha già fatto la trasformazione dei tipi - non sono tutte stringhe da ritrasformare in int, float, etc.
for f in features:
    data.append({
        'title': f['properties']['title'],
        'impact': f['properties']['sig'],
        'magnitude': f['properties']['mag'],
        'location': f['properties']['place'],
        'latitude': f['geometry']['coordinates'][0],
        'longitude': f['geometry']['coordinates'][1]
    })
    
print(data)
```

```python
import folium

m = folium.Map(location=[-122, 38], zoomstart=-1000)
datapoints = folium.map.FeatureGroup()
for posti in data:
    for k,v in posti.items():
        if k=='latitude':
            lat=v
        if k=='longitude':
            lon=v
    print(lat,lon)
    datapoints.add_child(
           folium.vector_layers.CircleMarker(
            [lat, lon],
            radius=5, # set the markers width
            color='green',
            fill=True,
            fill_color='white',
            fill_opacity=.2
        )
    )
m.add_child(datapoints)
m
```
